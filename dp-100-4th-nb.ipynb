# LAB 6 -- Feature Engineering and Scaling

-- Prepare data for ML models
-- Split Features and Target
-- Apply SCaling

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.preprocessing import StandardScaler
    
```

```python
import pandas as pd
```

```python
df = pd.read_csv('diabetes.csv')
```

```python
df.info()
```

```python
X = df.drop('Diabetic', axis=1)

# X -- Represents all input features (except target column (Diabetic))

Y = df['Diabetic']

# Y -- The Target Variable (what we are going to predict)
```

```python
scaler = StandardScaler()

# StandardScaler --  Mean = 0,  SD = 1

# Many ML models (Logistic Regression, KNN algo) 

# Example  -- Age (20-80) vs Glucose (0-200)  
```

```python
x_scaled = scaler.fit_transform(X)
```

```python
x_train, X_test, y_train, Y_test = train_test_split(x_scaled, Y, test_size=0.20)

# Training data  -- 80%  (used to train model)

# Test Set -- 20 %  (Used to Evaluate the performance)
```

# LAB 7  -- Machine Learning (Multiple Algorithms)


```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
models = {
         "Logistic" : LogisticRegression(),
         "Decision Tree" : DecisionTreeClassifier(),
         "Random Forest" : RandomForestClassifier()
}
```

```python
for name, model in models.items():
    model.fit(x_train, y_train)
    print(f"{name} Accuracy: ", model.score(X_test, Y_test))
```
